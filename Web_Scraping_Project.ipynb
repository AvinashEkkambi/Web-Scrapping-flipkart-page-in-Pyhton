```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.flipkart.com/hero-xtreme-125r-ibs-booking-ex-showroom-price/p/itm703abeab01d68?pid=MCYGYH9PQ55WUD94&marketplace=FLIPKART'
```

```python
url
```

```python
page=requests.get(url)
print(page)
```

```python
data=page.text
```

```python
soup=BeautifulSoup(data,'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
title=soup.title
print(title.text)
```

```python
price= soup.find('div',class_="Nx9bqj CxhGGd yKS4la")
print(price.text)
```

```python
highlights=soup.find('div',class_='xFVion')
print(highlights.text)
```

```python
avgrating=soup.find('div', class_="XQDdHH")
print(avgrating.text)
```

```python
allratings=soup.findAll('div',class_="XQDdHH Ga3i8K")
rates=[]
for i in allratings:
    rates.append(i.text)
print(rates)
```

```python
customernames=soup.findAll('p',class_="_2NsDsF AwS1CA")
names=[]
for i in customernames:
    names.append(i.text)
names
```

```python
que=soup.findAll('div',class_='wys2hv _43gOsC')
ques=[]
for i in que:
    ques.append(i.text)
ques
```

```python
bankoff=soup.findAll('div',class_='f+WmCe')
bank=[]
for i in bankoff:
    bank.append(i.text)
bank
```

```python
general=soup.findAll('div',class_='_1OjC5I pKBIUH')
gen=[]
for i in general:
    gen.append(i.text)
gen
```

```python
comm=soup.findAll('p',class_='z9E0IG')
comms=[]
for i in comm:
    comms.append(i.text)
comms
```

```python
df=pd.DataFrame()
df['Custmer Name']=names
df['Comments']=comms
df['Ratings']=rates
df
```
